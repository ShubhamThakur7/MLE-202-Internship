```python
wordnet = WordNetLemmatizer()
```

```python
headlines=[]

for i in range(0,len(news_data['Headlines'])):
    lines = re.sub('[^a-zA-Z]',' ',news_data['Headlines'][i])
    lines = lines.lower()
    lines = lines.split()
    
    lines = [wordnet.lemmatize(word) for word in lines if word not in set(stopwords.words('english'))]
    lines = ' '.join(lines)
    headlines.append(lines)
```
